## 5. Mezclando valores de tipos distintos

```python
vector = np.array([1,2,3,4,5])
vector
```

```python
type(vector[0])
```

```python
# Mezcla de enteros con flotantes

vector = np.array([1,2,3.0,4,5])
vector # Ha convertido todos los enteros a flotantes
```

```python
# Mezcla de booleanos con enteros

vector = np.array([1,2,3,4,False])
vector # Convierte los booleanos en 0 y 1 (false y true)
```

```python
# Mezcla de booleanos con flotantes

vector = np.array([1.0,2.0,3.2,4.0,True])
vector # Convierte los booleanos en 0. y 1. (false y true)
```

```python
# Mezcla de enteros con cadenas

vector = np.array([1,2,3,"Hola"])
vector # Convierte todo a cadenas
```

```python
# ¿Y si mezclamos todo? Usa lo más genérico

vector = np.array([1,2.5,3,"Hola", True])
vector # Convierte todo a cadenas
```
