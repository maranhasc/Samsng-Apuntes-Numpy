```python
df_iris.plot.scatter(x='sepal_length', y ='sepal_width')
plt.show()
```

```python
diccionary_colores={"setosa": "red","virginica": "green","versicolor": "blue"}
#en el cambio de nombre del principio species se llama variante
colores=df_iris["Variante"].apply(lambda x:diccionary_colores[x])
print(colores)
```

```python
df_iris.plot.scatter(x='sepal_length', y='sepal_width', c=colores)
plt.show()
```

```python
df_iris.plot.scatter(x='petal_length', y='petal_width', c=colores)
plt.show()
```
