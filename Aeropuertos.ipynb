```python
#Ej1 Buscar los aeropuertos españoles  iso_county ="ES" agruparlos por tipo de aeropuerto y contarlos y que aparezcan ordenados 
# de forma descendente por numero Fichero airport.cvs, aero
```

```python
#Paso1 buscar aeropuertos de España
aero_espa= airports[airports.iso_country=="ES"]
aero_espa
```

```python
#Agrupa por tipo y cuenta los que hay de cada tipo
result1= aero_espa.groupby(["type"]).count()["id"]
result1
```

```python
#muestra el resultado ordenado descendente
result1.sort_values(ascending=False)
```

```python
#Todas las sentencias agurpadas busca los aeropuertos españoles, agrupados por tipo y cuentalos, muestra el resultado ordenado
#descendente
aero_espa[airports.iso_country == "ES"]\
     .groupby(["type"]).count()["id"]\
     .sort_values(ascending=False)

```

```python
#Todas las sentencias agurpadas busca los aeropuertos españoles, agrupados por tipo y cuentalos, muestra el resultado ordenado
#descendente
aero_espa[airports.iso_country == "ES"].groupby(["type"]).count()["id"].sort_values(ascending=False)
```

# Ejercicio 2
### Buscar los cinco paises que tienen más aeropuertos de tipo "large_airport" y 
### mostrarlos por pantalla ordenados en orden descendente por número, 
### pero hay que mostrar el nombre completo del pais, no su código iso-country (ficheros: airports.csv y
### countries.csv).


```python
merge_data = pd.merge(airports, paises, left_on="iso_country", right_on="code", how="inner")
merge_data
```

```python
merge_data.info()
#los campos que estan en los dos sitios los pone porque se llaman igual uno es el nombre del aeropuesto 
#y otro el nombre del pais
```

```python
#name_x es el del aeropuerto y name_y el del pais
#filtro por aeropuertos grandes o largos
large_airports = merge_data[merge_data["type"]=="large_airport"]
large_airports
#merge_data.gropuby("name_y")large_airports.
```

```python
#con los parentesis toma todo lo que hay dnetro como una sentencia
#agrupo por nombre de pais, contamos, ordenamos descendiente y cogemos los cinco primeros
(large_airports.groupby("name_y")["id_x"]
     .count()
     .sort_values(ascending = False)
     .head(5))
```

```python
#Todo el ejercicio 2 junto

(pd.merge(airports, paises, left_on="iso_country", right_on="code", how="inner")[merge_data["type"]=="large_airport"]
     .groupby("name_y")["id_x"]
     .count()
     .sort_values(ascending = False)
     .head(5))
```
