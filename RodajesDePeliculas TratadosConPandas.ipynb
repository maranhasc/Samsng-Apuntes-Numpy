<h3> 12.-Opcional: plantea al menos tres operaciones con el dataframe que consideres que son particularmente interesantes.
 </h3>

```python
#Busca todas las peliculas de la distribuidora "Columbia Pictures" y agrupalas por año y cuenta las peliculas de cada año 
#ordenadas descendentemente
Distribuidora_columbia=df_rodajes[df_rodajes.Distributor == "Columbia Pictures"]
Resultado=Distribuidora_columbia.groupby(['Release Year']).count()["Title"].sort_values(ascending=False)
Resultado
```

```python
#Agrupa por director y distribuidora y muestra el nº de peliculas
agrupacion = df_rodajes.groupby(["Director","Distributor"])["Title"].count()
agrupacion
```

```python
#Agrupa por director y actor y  y muestra el nº de peliculas enlas que han trabajado juntos
agrupacion_actor = df_rodajes.groupby(["Director","Actor 1"])["Title"].count()
agrupacion_actor
```
