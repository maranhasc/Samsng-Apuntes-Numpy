### mezclar (combinar o fusionar) dataframes

```python

dataA = {"Nombre": ["Happy Potter", "David Baker", "John Smith", "Juan Martínez", "Jane Connor"], "Gender": ["Male", "Male", "Male", "Male", "Female"], "Age": [23, 32, 22, 36, 30]}
grupoA = pd.DataFrame(dataA)
grupoA
```

```python

dataB = {"Nombre": ["John Smith", "AlexDu Bois", "Joanne Rowling", "Jane Connor"], "Position": ["Intern", "Team Lead", "Manager", "Manager"], "Wage" : [25000, 75000, 90000, 70000]}
grupoB = pd.DataFrame(dataB)
grupoB
```

```python
#Inner join coge los comunes a los dos
pd.merge(grupoA, grupoB, on ="Nombre", how="inner")
```

```python
#Left join coge todoslos de la izquierda y de la derecha los comunes
pd.merge(grupoA, grupoB, on ="Nombre", how="left")
```

```python
#right join coge todos los de la dercha y de la izquierda solo los comunes
pd.merge(grupoA, grupoB, on ="Nombre", how="right")
```

```python
#outer join coge los dos
pd.merge(grupoA, grupoB, on ="Nombre", how="outer")
```

```python
grupoB.rename(columns = {"Nombre":"Name"}, inplace= True)
grupoB
```

```python
pd.merge(grupoA, grupoB, left_on="Nombre", right_on="Name" how="inner")
```

```python
pd.concat([grupoA, grupoB], axis=1)
```
