```python
import folium
import pandas as pd
```

```python
# traer datos directamente de una url
dataframe = pd.read_csv("Datos/ocupappublicosmun.csv", sep = ",", header="infer")
dataframe
```

```python
longitude = dataframe["longitude"].head(5)
longitude
```

```python

latitude = dataframe["latitude"].head(5)
latitude
```

```python

map = folium.Map(location=(latitude[1], longitude[1]), zoom_start=15)
map
```

```python
for i in range(len(latitude)):
    folium.Marker(
    location=[latitude[i], longitude[i]],
    icon=folium.Icon(icon="cloud"),
).add_to(map)
    
```

```python
#Cobisa
m = folium.Map(location = [39.8044211, -4.0247277], zoom_start = 12)

# Círculo con un radio de 5km
folium.Circle(
    location = [39.8044211, -4.0247277],
    radius = 5000,
    color = "red",
    fill = True,
    fill_color = "blue").add_to(m)

m
```

```python
m = folium.Map(location = [39.8044211, -4.0247277], zoom_start = 18)

# Agregar un marcador con icono al mapa
folium.Marker(location = [39.8044211, -4.0247277],
              icon = folium.Icon(icon = "cloud", color = "green")).add_to(m)

m 
```

```python
# Coordenadas de cobisa
coords = pd.DataFrame({'lat': [-4.0191098, -4.0200226, -4.0247277],
                       'lon': [39.8065881 , 39.8083123, 39.8044211]})

m = folium.Map(location = [39.8044211, -4.0247277], zoom_start = 15)

# Agregar varios marcadores al mapa
for index, row in coords.iterrows():
  folium.Marker(location = [row["lon"], row["lat"]]).add_to(m)

m 
 
```
