## Indices

```python
import pandas as pd
import numpy as np
```

```python
data = {"Nombre": ["Luis", "Josefa", "Carlos", "Lucía"], "Edad": [20, 25, 40, 51], "EstadoCivil": ["S", "C", "V", "S"], "Altura": [1.78, 1.62, 1.90, 1.72], "FechaIngreso": ['3:12:2000', '23:02:2012', '7:08:1999', '03:10:2013']}
empleados = pd.DataFrame(data)
empleados
```

Como no se ha indicado el indice, pone por defecto de 0 a ....n

```python
empleados = pd.DataFrame(data,index=[2,5,2,6])
empleados
```

```python
empleados.loc[2]  #Los datos con indice 2
```

```python
empleados.iloc[2]  #Los datos con indice 2 po DEFECTO
```

```python
# Una columna se puede convertir en indice
empleados.set_index("Nombre",inplace=True) #Nombre es el nuevo indice
```

```python
empleados
```

```python
empleados.index
```

```python
empleados.loc["Lucía"]  #Serie
```

```python
#Volver a poner el indice por defecto
empleados.reset_index()  #empleados.reset_index(inplace=True) para hacerlo definitivo
```

```python
empleados.transpose() #Filas por columnas del dataframe
```

```python
empleados.reset_index(inplace=True)  #Reseteo definitivo
```

```python
empleados
```

```python
#Indices a 2 niveles
empleados.set_index(["Nombre","Edad"], inplace=True)
empleados
```

## Indices multiples

```python
cabecera=["Altura","Anchura","Profundidad"]
indice_exterior=["Grupo1"]*3+["Grupo2"]*3
print(indice_exterior)

indice_interior=[1,2,3]*2
print(indice_interior)

indice_multiple=list(zip(indice_exterior,indice_interior)) #Lista de tuplas
indice_multiple
```

```python
indice=pd.MultiIndex.from_tuples(indice_multiple)
indice
```

```python
dataframe=pd.DataFrame( data=np.random.randn(6,3), columns=cabecera, index=indice)
dataframe
```
